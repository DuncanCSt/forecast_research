# Generate data of varying degrees of state switching and evaluate model fits

```python
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt
```

```python

means = [[3, 1], [5, 1], [7, 1]]
densities = [0.3, 0.4, 0.3]

for sd in [0.5, 0.7, 1.0, 1.5]:
    x_vals = np.linspace(0, 10, 100)
    pdf = [np.sum([densities[i] * norm.pdf(x, loc=means[i][0], scale=sd) for i in range(len(means))]) for x in x_vals]
    plt.plot(x_vals, pdf, label=f'SD={sd}')

plt.legend()
```

```python
for mean in [0.5, 1.5, 2]:
    means = [5-mean, 5, 5+mean]
    x_vals = np.linspace(0, 10, 100)
    pdf = [np.sum([densities[i] * norm.pdf(x, loc=means[i], scale=1) for i in range(len(means))]) for x in x_vals]
    plt.plot(x_vals, pdf, label=f'Mean offset={mean}')

plt.legend()
```
